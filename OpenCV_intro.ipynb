```
# Biblioteca OpenCv no Python
import cv2
```

```
#Abrir a imagem do ambiente 

imagem = cv2.imread("/content/drive/My Drive/Colab Notebooks/Visão Computacional/images/Lenna.png")

#Tipo de dado
print(type(imagem))

#Formato do dado
# 3 dimensões porque nossa imagem é colorida
print(imagem.shape)
```

```
#Converter imagem para escala de cinza
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

#Tipo de dado
print(type(imagem_gray))

#Formato do dado
print(imagem_gray.shape)
```

```
#Mostrar imagem com o openCV

#cv2.show( "Imagem de saída", imagem_gray )

from google.colab.patches import cv2_imshow
cv2_imshow( imagem_gray ) 
```

```
print(imagem_gray)
for linha in range(len(imagem_gray)):
  for coluna in range(len(imagem_gray[0])):
    imagem_gray[linha,coluna] /= 2

print( imagem_gray )
cv2_imshow( imagem_gray ) 
# entre a imagem mais clara e imagem mais escura sempre prefira a mais escura, pois a mais clara está perto do limite das cores
```
